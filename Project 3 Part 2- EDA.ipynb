```python
# Budget filter for values greater than 0
budget_filter = df['budget'] > 0

df[budget_filter].info()
```

- 546 movies listed for budget values greater than 0.

```python
# Revenue filter for values greater than 0
revenue_filter = df['revenue'] > 0

df[revenue_filter].info()
```

```python
# Combine filters to exclude 0 values 
filters_combined = budget_filter & revenue_filter
df[filters_combined].info()
```

2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
df[filters_combined]['certification'].value_counts()
```

```python
#Count plot
ax = sns.countplot(data = df[filters_combined], x = 'certification')
# this is for the title of plot
ax.set_title('Counts for Certification Categories');
```

3. What is the average revenue per certification category?

```python
avg_rev = df[filters_combined].groupby('certification')['revenue'].mean()

avg_rev
```

```python
# Using seaborn to plot the group means
ax = sns.barplot(data=df[filters_combined], x='certification', y="revenue")
```

```python
# Plot the average of each as a bar
fig, ax = plt.subplots()
ax.bar(avg_rev.index, avg_rev.values);
ax.set_ylabel('revenue')
ax.set_xlabel('certification')
ax.set_title('Average Revenue by Certification');
```
